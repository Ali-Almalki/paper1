```python
import matplotlib.pyplot as plt

# Data
datasets = ["Dataset5", "Dataset6", "Dataset7", "Dataset8", "Dataset9"]
training_loss = [0.732000,0.230000,0.200000,0.300000,0.654000]
validation_loss = [0.430000,0.200000,0.190000,0.290000,0.440000]
accuracy = [0.965000,1.000000,1.000000,1.000000,0.969000]

# Create larger figure
fig, ax = plt.subplots(figsize=(8,6))

# Set bar width
bar_width = 0.3


# Position bars
r1 = range(len(datasets))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Plot bars
ax.bar(r1, training_loss, color='#64CCC5', width=bar_width, label='Training Loss')
ax.bar(r2, validation_loss, color='#176B87', width=bar_width, label='Validation Loss')
ax.bar(r3, accuracy, color='#04364A', width=bar_width, label='Accuracy')

# Labels and title
ax.set_xticks([r + bar_width for r in range(len(datasets))])
ax.set_xticklabels(datasets)
ax.set_ylabel('Value')


# Add value labels
for i, v in enumerate(training_loss):
    ax.text(i, v, str(v), ha='center', va='bottom')
for i, v in enumerate(validation_loss):
    ax.text(i + bar_width, v, str(v), ha='center', va='bottom')
for i, v in enumerate(accuracy):
    ax.text(i + 2*bar_width, v, str(v), ha='center', va='bottom')

# Add legend and tight layout
ax.legend()
plt.tight_layout()

# Show plot
plt.show()
```
